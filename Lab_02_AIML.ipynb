```python
def bfs(graph,source):
  Q=Queue()
  visited_vertices=set()
  Q.put(source)
  visited_vertices.update({source})
  while not Q.empty():
    vertex=Q.get()
    print(vertex,end="-->")
    for vertices in graph[vertex]:
      if vertices not in visited_vertices:
        Q.put(vertices)
        visited_vertices.update({vertices})
print("BFS traversal of graph with Source 0 is:")
bfs(graph,0)
```

#Part-02:Implement Depth First Search Algorithm

image.png

```python
graph1={"A":["C","D","G"],"B":["A"],"C":["A","D","F"],"D":["A","C","E","G","H"],"E":["D","F","H"],"F":["C","E"],"G":["A","D","H"],"H":["D","E","G"]}
print("The adjaceny List representing the graph is:")
print(graph1)
```

```python
def dfs(graph, node, visited):
    if node not in visited:
        visited.append(node)
        for k in graph[node]:
            dfs(graph, k, visited)
    return visited

visited = dfs(graph1, 'A', [])
print(visited)
```
